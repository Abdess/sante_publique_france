## Fonctions

Fonctions commune

```python
# Calculer le nombre de lignes d'un dataframe spécifique
def get_rows(df):
    rows = df.shape[0]
    return rows


# Retourner les 10 premiers enregistrements par nom de colonne
def get_top_10(df, column_name):
    df_result = df.groupby([
        column_name
    ]).size().reset_index(name='Nombre').sort_values('Nombre',
                                                     ascending=False).head(10)

    rows = get_rows(df)
    df_result['Pourcentage'] = df_result['Nombre'] / rows * 100

    return df_result.copy()


# Compter le nombre de lignes vides pour un dataframe spécifique.
def get_empty_rows(df, column_name):
    return df.query(str(column_name) + '.isnull()').count()


# Afficher le graphique en barre avec le nombre
def show_plot(df, column_name):
    df_plot = df.loc[:, [column_name, 'Nombre']].set_index(
        column_name)['Nombre'].copy()

    plt.figure(figsize=(15, 10))

    sns.set(style="whitegrid")
    g = sns.barplot(df_plot, df_plot.index, orient='h')

    plt.title('Les plus grands ' + column_name, fontsize=20)
    plt.xlabel("Nombre d'occurrences", fontsize=14)
    plt.ylabel(column_name, fontsize=14)

    # Afficher le nombre à droite du graphique à barres
    for i, v in enumerate(df_plot.values.tolist()):
        g.text(v + 3, i + .25, str(v), color='black', fontweight='normal')

    plt.show()


# Imprimer les principaux éléments de statistiques
def print_stats(df, column_name):
    total_rows = get_rows(df)
    empty_rows = str(df[column_name].isna().sum())
    perc = int(empty_rows) * 100 / int(total_rows)
    print("Nombre d'enregistrements vides pour " + str(column_name) + ': ' +
          str(empty_rows) + ' (' + str(perc) + '%)')
    print(df[column_name].agg(['nunique', 'count', 'size']))
```

### Aperçu du moment où les produits ont été créés par les contributeurs

```python
top_5_creators = df.loc[(df['creator'] == 'kiliweb') |
                        (df['creator'] == 'usda-ndb-import') |
                        (df['creator'] == 'openfoodfacts-contributors') |
                        (df['creator'] == 'org-database-usda') |
                        (df['creator'] == 'elcoco')][[
                            'creator', 'created_datetime'
                        ]]

top_5_creators['mod_datetime'] = top_5_creators['created_datetime'].apply(
    lambda x: x[0:4])
top_5_creators.drop('created_datetime', axis=1)

mod_pivot = pd.pivot_table(top_5_creators,
                           index=top_5_creators['mod_datetime'],
                           columns='creator',
                           aggfunc=len,
                           fill_value=0)
mod_pivot.plot(kind="line",
               figsize=(12, 8),
               title="Quand les produits ont-ils été créés ?")
mod_pivot.plot(kind="bar",
               figsize=(12, 8),
               title="Quand les produits ont-ils été créés ?")
```

Jusqu'à présent, nous pouvons voir que Kiliweb est le contributeur le plus actif et ajoute continuellement des produits. C'est le contributeur le plus actif. La plupart des produits sont donc probablement principalement de France. 
Nous pouvons déjà supposer, sur la base des principaux contributeurs, que les produits français prédominent.

usda-ndb-import, qui est le 'United States Department of Agriculture' semble ne plus contribuer à la base de données open food facts et n'a été importé qu'une seule fois.

La communauté ouverte des contributeurs d'openfoodfacts semble s'agrandir au fil des années. La diversité des contributeurs peut conduire à des données non concordantes.

```python
# Voici une fonction pour faciliter l'exploration :
# Obtenir les variables pour lesquelles le pourcentage
# des valeurs manquantes se situe dans la fourchette basse et haute des %.
def get_variables_with_nan_in_range(data, low, high, dtypes):
    if high >= 100:
        high = 100.1

    nan_counts = data.select_dtypes(dtypes).isnull().sum()
    nan_counts_percent = nan_counts / data.shape[0] * 100

    filter = (nan_counts_percent >= low) & (nan_counts_percent < high)
    nan_counts = nan_counts[filter]
    nan_counts_percent = nan_counts_percent[filter]
    counts = data.shape[0] - nan_counts
    counts_percent = 100 - nan_counts_percent

    return pd.DataFrame({
        'nombre': counts,
        'nombre %': counts_percent,
        'nombre de NaN': nan_counts,
        'nombre de NaN %': nan_counts_percent
    })


# Données alimentaires avec NaN
get_variables_with_nan_in_range(df_fr, 0, 95, 'number')
```

Dans la fourchette 0-95% de valeurs manquantes, nous avons des variables pour les macronutriments (glucides/lipides/protéines) et le premier niveau des minéraux et vitamines les plus courants.
```python
# Obtenir les principales occurrences
def get_top_occurencies(dfallergens, cols, n=10):
    count = dfallergens.groupby(cols).size().sort_values(ascending=False)
    count_percent = round(count / dfallergens.shape[0] * 100, 2)
    return pd.DataFrame({'count': count, 'count %': count_percent}).head(n)


# Obtenir les occurrences les plus importantes contenant
# une expression rationnelle ou une chaîne de caractères.
def get_top_occurencies_containing(dfallergens, col, value, n=10):
    occurences = get_top_occurencies(dfallergens, col, dfallergens.shape[0])
    return occurences[occurences.index.str.contains(
        r'{}'.format(value))].head(n)
```

```python
# 1. Extraire tous les états (retourne une série multi-indexée
# avec les états extraits comme valeur, <code + match> | state)
# Lors de l'extraction, nous donnons le nom "états" au groupe extrait
# Note : Nous aurions pû aussi utiliser : pd.Series.str.get_dummies(sep=',')
df_states = df_fr.states_fr.str.extractall(r"(?P<états>[^,]+)")
df_states.head(8)
```

### Construire un dictionnaire des taxonomies des allergènes

Nos données sont maintenant prêtes à être traduites, nous devons maintenant construire un dictionnaire pour traduire les allergènes et remplacer les différents synonymes par leurs taxonomies correspondantes. Le GitHub d'OpenFoodFact fournit les taxonomies à cette url : https://github.com/openfoodfacts/openfoodfacts-server/tree/main/taxonomies

Sur cette page, on peut voir les différentes taxonomies : fr:gluten, fr:crustacés, fr:œufs, etc. avec différents synonymes dans de nombreuses langues différentes.

Note : Il nous est aussi possible de reconstruire les allergènes grâce aux ingrédients à l'aide du fichier "allergens.result.txt"

```python
# Récupération du fichier contenant les taxonomies d'allergènes
taxonomies = requests.get(
    'https://raw.githubusercontent.com/openfoodfacts/openfoodfacts-server/main/taxonomies/allergens.txt'
)
taxonomies = str(taxonomies.text)
```

Les différentes taxonomies d'allergènes sont séparées par des lignes vides.
Nous pouvons construire une liste en séparant le texte avec ce séparateur, et construire un dataframe à partir de la liste.

```python
# Supprimer le flag "stopwords" au début du fichier
taxonomies = re.sub(r'stopwords.*?\n', '', taxonomies)

# Supprimer l'annotation wikidata à la fin de chaque taxonomie
taxonomies = re.sub(r'wikidata.*', '', taxonomies)

# Supprimer les lignes vides
taxonomies = re.sub(r'^(?:[\t ]*(?:\r?\n|\r))+', '\n', taxonomies)

# Supprimer les commentaires
taxonomies = re.sub(r'\n#.*', '', taxonomies)

# Supprimez les espaces après la virgule.
taxonomies = re.sub(r',\s+', ',', taxonomies)

# Retirer les 3 retours à la ligne successif
taxonomies = re.sub("([ \t]*\n){3,}", "\n\n", taxonomies)

# Retirer les espaces
taxonomies = taxonomies.strip()

# Tout en minuscules
taxonomies = taxonomies.lower()

# Les allergènes avec leurs synonymes et leurs traductions sont séparés par une ligne vide.
taxonomies = taxonomies.split('\n\n')
```

```python
# Construire un dataframe
taxonomies = pd.DataFrame({'text': taxonomies})
taxonomies.head()
```

La taxonomie est le premier mot après le drapeau "fr" sur chaque observation. Extrayons-les et remplaçons l'index par ceux-ci

```python
# Extraire la taxonomie principale des allergènes
# du premier mot français jusqu'à la virgule ou le retour à la ligne
taxonomies['allergen'] = taxonomies.text.str.extract(r'fr:([^\n,]+)')

# Définir l'allergène comme index du dataframe.
taxonomies.set_index('allergen', inplace=True)

taxonomies.head()
```

Nous devons maintenant extraire chaque langue pour avoir une ligne par allergène et par langue. Les langues sont séparées par un retour à la ligne

```python
# Extract all langages (to have one row per allergen and langages)
allergens_translations = taxonomies.text.str.extractall(
    r'(?P<translations>[^\n]+)')

# Reset index
allergens_translations.reset_index(inplace=True)

# Drop the automatically created match column
allergens_translations.drop(['match'], axis=1, inplace=True)

allergens_translations.head()
```

### Traduction des allergènes
Nous avons maintenant tout pour remplacer les données sources des allergènes par leurs taxonomies correspondantes :

```python
# Fonction qui renvoie la taxonomie de l'allergène
# si elle est trouvée ou sinon l'allergène déjà existant
def get_allergen_translation(x):
    result = allergens_translations[
        allergens_translations.translations.str.contains(x)]
    if result.shape[0] > 0:
        return result.allergen.iloc[0]
    else:
        return x


# Fonction qui renvoie si une taxonomie d'allergène a été trouvée
def has_translation(x):
    result = allergens_translations[
        allergens_translations.translations.str.contains(x)]
    return result.shape[0] > 0


# Ajouter une variable avec la traduction/taxonomie des allergènes
df_allergens['allergen_translation'] = df_allergens.allergen.map(
    get_allergen_translation)

# Ajouter une variable qui indique si une traduction/taxonomie a été trouvée
df_allergens['translated'] = df_allergens.allergen.map(has_translation)
```

Comme nous pouvons avoir plusieurs synonymes pour la même taxonomie d'allergènes, nous avons probablement de nouveaux doublons après la traduction. Nous devons les supprimer :

```python
# Réinitialiser l'index
df_allergens.reset_index(inplace=True)

# Supprimer les doublons créés par la traduction parce
# qu'un produit a plusieurs synonymes pour le même allergène
df_allergens.drop_duplicates(subset=['code', 'allergen_translation'],
                             inplace=True)

# Capitaliser la traduction de l'allergène
df_allergens.allergen_translation = df_allergens.allergen_translation.str.capitalize(
)

df_allergens.loc[:,
                 ['code', 'product_name', 'allergen', 'allergen_translation'
                  ]].head(10)
```

### Sélection des variables et des observations

```python
# Variables que nous voulons conserver
# macronutriments de niveau supérieur (ex : protéines, lipides, glucides)
# et leur premier niveau de sous catégorie
# (ex : Lipides = graisse saturée, monoinsaturés, ...)
col_list = [
    'product_name', 'pnns_groups_1', 'pnns_groups_2', 'source', 'energy_100g',
    'carbohydrates_100g', 'sugars_100g', 'starch_100g', 'polyols_100g',
    'fat_100g', 'saturated-fat_100g', 'monounsaturated-fat_100g',
    'polyunsaturated-fat_100g', 'trans-fat_100g', 'proteins_100g',
    'casein_100g', 'serum-proteins_100g', 'fiber_100g', 'salt_100g'
]

# Variables qui seront utilisées pour l'analyse
# uniquement les macronutriments de niveau supérieur et l'énergie
facts_for_analysis = [
    'energy_100g', 'carbohydrates_100g', 'fat_100g', 'proteins_100g',
    'fiber_100g', 'salt_100g'
]

# Conserver uniquement les variables requises et
# exclure les groupes de PNNS inconnus (Unknown)
df_nutrients = df_nutrients[df_nutrients.pnns_groups_1 != 'Unknown'].reindex(
    columns=col_list)

# Supprimer les observations avec des valeurs
# manquantes en macronutriments et en énergie
df_nutrients.dropna(subset=facts_for_analysis, inplace=True)

# Remplacer les valeurs NaN par 0
df_nutrients.fillna(value=0, inplace=True)

print("Nombre d'observations : {}".format(df_nutrients.shape[0]))
df_nutrients.head()
```

### Suppression des produits dont les valeurs des sous-ensembles sont incohérentes

Rappelez-vous que nous avons déjà éliminé les produits dont les faits sont négatifs ou supérieurs à 100g. Ici, nous allons simplement supprimer les produits qui ont des valeurs de faits incohérentes entre le niveau supérieur et leur sous-ensembles.

```python
# Fonction permettant de supprimer les observations incohérentes
# ATTENTION : cette fonction fonctionne (😊) car les variables
# sont bien ordonnées.
def drop_observations_with_incoherent_breakdown(df, col_total,
                                                col_breakdown_start,
                                                col_breakdown_end, epsilon):
    breakdown_total = df.loc[:,
                             col_breakdown_start:col_breakdown_end].sum(axis=1)
    condition_filter = breakdown_total > (df[col_total] * (1 + epsilon))
    df.drop(df[condition_filter].index, inplace=True)


# tolérance pour la mesure des erreurs d'arrondi et des erreurs de protocole
epsilon = 0.1

# Glucides : supprimer les observations dont la somme de la répartition
# des glucides est supérieure au total des glucides (avec tolérance).
drop_observations_with_incoherent_breakdown(df_nutrients,
                                            col_total='carbohydrates_100g',
                                            col_breakdown_start='sugars_100g',
                                            col_breakdown_end='polyols_100g',
                                            epsilon=epsilon)

# Lipides : supprimer les observations dont la somme de la répartition
# des lipides est supérieure au total des lipides (avec tolérance).
drop_observations_with_incoherent_breakdown(
    df_nutrients,
    col_total='fat_100g',
    col_breakdown_start='saturated-fat_100g',
    col_breakdown_end='trans-fat_100g',
    epsilon=epsilon)

# Protéines : supprimer les observations dont la somme de la répartition
# des protéines est supérieure au total des protéines (avec tolérance).
drop_observations_with_incoherent_breakdown(
    df_nutrients,
    col_total='proteins_100g',
    col_breakdown_start='casein_100g',
    col_breakdown_end='serum-proteins_100g',
    epsilon=epsilon)

print("Nombre d'observations : {}".format(df_nutrients.shape[0]))
```

### Suppression des valeurs aberrantes par groupes de PNNS

Dans cette section, nous allons supprimer les produits aberrants dans chaque groupe PNNS. Un produit d'un groupe PNNS est considéré comme aberrant si l'une des valeurs suivantes est éloignée de plus de 3 sigma de la moyenne de son groupe PNNS :

- energy_100g
- carbohydrates_100g
- fats_100g
- proteins_100g
- fiber_100g
- salt_100g

```python
# Fonction pour éliminer les valeurs aberrantes en ce qui concerne les
# variables de haut niveau des faits (définies dans facts_for_analysis)
def drop_outliers(group, nsigma=3):
    g = group.loc[:, facts_for_analysis]
    is_outlier = np.abs(g - g.mean()) > (nsigma * g.std())
    outliers = g[is_outlier.any(axis=1)]
    return group.drop(outliers.index, axis=0)


# Groupe de produits par groupes de PNNS
by_pnns = df_nutrients.groupby(['pnns_groups_1', 'pnns_groups_2'])

# Supprimer les valeurs aberrantes
df_nutrients = by_pnns.apply(drop_outliers)

# Remise à zéro de l'index car après le groupby, il s'agit
# d'un index multiple : pnns_groups_1 | pnns_groups_2 | row index
df_nutrients.reset_index(level=[0, 1], drop=True, inplace=True)

print("Nombre d'observations : {}".format(df_nutrients.shape[0]))
```

Une fois les valeurs aberrantes supprimées, vérifions le nombre de produits par groupe PNNS.

Il peut être tentant de comparer les valeurs des moyennes et d'en tirer des conclusions faciles et rapides, mais ce n'est jamais une bonne idée de comparer des moyennes sans connaître leur écart-type. 1. Quels sont les aliments qui contiennent le plus de fibres, par ordre d'importance ?

Tout d'abord, nous pouvons constater que le classement serait différent en comparant les médianes; les noix seraient l'aliment qui contient le plus de fibres. Mais la bonne question à se poser est de savoir si les différences moyennes/médianes sont significatives ? Peut-on vraiment dire que les noix contiennent plus de fibres que les légumineuses, ou que les fruits secs ?

Le test de Student, ou test t, peut nous aider à répondre à ces questions. L'hypothèse nulle consiste à dire que les moyennes sont les mêmes et que la différence observée est due au hasard. Le test t nous indique si nous pouvons ou non rejeter l'hypothèse nulle.

```python
# Importer la méthode du test t
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html
from scipy.stats import ttest_ind


# Fonction permettant de calculer un test t
# indépendant avec  des résultats agréables à lire
# https://en.wikipedia.org/wiki/Student%27s_t-test#Independent_two-sample_t-test
def compute_ttest_ind(df, group_1, group_2, variable):
    stat, pvalue = ttest_ind(df.loc[df.pnns_groups_2 == group_1, variable],
                             df.loc[df.pnns_groups_2 == group_2, variable],
                             equal_var=False)
    print(
        "{} par rapport à {}\nstatistique: {}\np-value  : {}\nPeut-on rejeter l'hypothèse nulle ? : {}\n"
        .format(group_1, group_2, stat, pvalue, pvalue < 0.05))


compute_ttest_ind(df_nutrients_fiber, 'Nuts', 'Legumes', 'fiber_100g')
compute_ttest_ind(df_nutrients_fiber, 'Nuts', 'Dried fruits', 'fiber_100g')
compute_ttest_ind(df_nutrients_fiber, 'Nuts', 'Breakfast cereals',
                  'fiber_100g')
compute_ttest_ind(df_nutrients_fiber, 'Dried fruits', 'Breakfast cereals',
                  'fiber_100g')
```

Nous parvenons à rejeter l'hypothèse nulle en comparant les noix et les légumineuses. Nous pouvons dire que les fruits à coque contiennent plus de fibres que les légumineuses, il serait donc logique d'établir un classement.

Comparons la composition en fibres du groupe "Fruits et légumes".

```python
# Construire un Dataframe avec seulement les produits
# du groupe 1 des PNNS "Fruits and vegetables".
df_nutrients_fiber = df_nutrients.loc[df_nutrients.pnns_groups_1 ==
                                      'Fruits and vegetables']

# Traçer une boîte à moustache pour comparer
# les statistiques de chaque groupe.
sns.set(style="whitegrid")

plt.figure(figsize=(16, 8))
sns.boxplot(x='pnns_groups_2',
            y='fiber_100g',
            data=df_nutrients_fiber,
            order=['Dried fruits', 'Vegetables', 'Fruits', 'Soups'])

plt.xlabel('Groupe PNNS')
plt.ylabel('Fibre [gr]')
plt.title('Les fibres dans les fruits et légumes')
plt.show()
```

Ici, il ne fait aucun doute que la différence entre les fruits, les légumes et les fruits secs est significative.

```python
# Calculer un test t
compute_ttest_ind(df_nutrients_fiber, 'Dried fruits', 'Fruits', 'fiber_100g')
compute_ttest_ind(df_nutrients_fiber, 'Vegetables', 'Fruits', 'fiber_100g')
```

La valeur p (p-value) est très faible, ce qui indique que nous pouvons rejeter avec confiance l'hypothèse nulle.

Il y a quelques exceptions, par exemple, la teneur en vitamine C est considérablement réduite lorsque le fruit est séché.

### Corrélation entre l'énergie et la composition en macronutriments

Avec le dernier dataframe, nous pourrions répondre à de nombreuses questions intéressantes, telles que les aliments les plus équilibrés, les aliments ayant le ratio le plus élevé de graisses, etc. Mais nous allons plutôt nous concentrer sur un sujet différent : l'énergie. Nous allons étudier la corrélation entre l'énergie et la composition des macronutriments et leur apport calorique respectif.

```python
x_vars = [
    'carbohydrates_100g', 'fat_100g', 'proteins_100g', 'fiber_100g',
    'salt_100g'
]

y_vars = ['energy_100g']

# Obtenir des nuage de points de l'énergie par rapport à
# chaque macronutriment pour tous les groupes d'aliments.
g = sns.pairplot(df_nutrients, kind='scatter', x_vars=x_vars, y_vars=y_vars)
g.fig.suptitle(
    'Énergie par rapport aux macronutriments (tous les groupes PNNS)')
g.set(xlim=(0, 100))
g.set(ylim=(0, 4000))
plt.show()
```

Nous pouvons constater une corrélation linéaire entre l'énergie et les lipides. Dans le nuage de points des glucides, nous pouvons également deviner une corrélation linéaire. Elle provient probablement d'un type d'aliment spécifique. Concernant les protéines, les fibres et le sel, il n'y a pas de corrélation visible évidente.

Dans le dataframe présentant la répartition des macronutriments par groupes de PNNS, on peut voir que :

1. Les produits gras (le groupe "Fats") ont tendance à avoir essentiellement des lipides.
2. Les fruits ont tendance à contenir essentiellement des glucides.
3. Il n'y a pas d'aliments qui contiennent essentiellement des protéines dans le même ordre que les fruits et les graisses, respectivement pour les glucides et les lipides.
4. Les légumineuses sont des aliments bien équilibrés et contiennent des glucides, des lipides et des protéines en bonne proportion.

Nous profiterons de la forte proportion de lipides et de glucides des produits gras et des fruits pour estimer leur apport calorique respectif. Ensuite, nous déduirons l'apport calorique en protéines des légumineuses en soustrayant à leur énergie connue l'énergie fournie par les graisses et les glucides.

```python
# Fonction pour tracer des nuages de points
# pour un groupe spécifique de PNNS 2
def pairplot_for_pnns_group(df,
                            pnns_group,
                            x_vars,
                            y_vars,
                            title,
                            xlim=(0, 100),
                            ylim=(0, 4000)):
    g = sns.pairplot(df[df.pnns_groups_2 == pnns_group],
                     kind='scatter',
                     x_vars=x_vars,
                     y_vars=y_vars)
    g.fig.suptitle(title)
    g.set(xlim=xlim)
    g.set(ylim=ylim)
    plt.show()


# Scatter plots of the energy versus each macronutrients for fruits only
pairplot_for_pnns_group(
    df_nutrients, 'Fruits', x_vars, y_vars,
    'Énergie par rapport aux Macronutriments (Fruits uniquement)')
```

Si l'on considère uniquement les fruits, on observe une corrélation linéaire claire entre l'énergie et les glucides. Nous pouvons calculer une régression linéaire. La pente nous donnera une approximation de l'apport calorique des glucides.

```python
# Importer la méthode de régression linéaire depuis scipy.stats
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html
from scipy.stats import linregress

# Conversion de kJ en kcal
kj_to_kcal = 0.239006

# Conserver uniquement les fruits
df_fruits = df_nutrients[df_nutrients.pnns_groups_2 == 'Fruits'].copy()

# Calculer la régression linéaire pour les fruits
# uniquement entre l'énergie et les glucides.
slope_carbohydrates, intercept_carbohydrates, r_value, p_value, std_err = linregress(
    x=df_fruits['carbohydrates_100g'], y=df_fruits['energy_100g'])

print("Équation de régression linéaire [kJ]    : Énergie = {} * Glucides + {}".
      format(slope_carbohydrates, intercept_carbohydrates))
print("Équation de régression linéaire [kcal]  : Énergie = {} * Glucides + {}".
      format(slope_carbohydrates * kj_to_kcal,
             intercept_carbohydrates * kj_to_kcal))
print("Énergie des Glucides [kJ]   apportent : {}".format(slope_carbohydrates))
print("Carbohydrates Energy [kcal] apportent : {}".format(slope_carbohydrates *
                                                          kj_to_kcal))
```

1 gramme de glucides fournit environ 4 kilocalories d'énergie.

Faisons de même pour les produits gras uniquement

```python
# Nuage de points de l'énergie en fonction de chaque
# macronutriment pour les produits gras uniquement.
pairplot_for_pnns_group(
    df_nutrients, 'Fats', x_vars, y_vars,
    'Énergie par rapport aux macronutriments (Produits gras uniquement)')
```

Nous avons une belle corrélation linéaire entre l'énergie et les lipides.

```python
# Conserver uniquement les produits gras
df_fats = df_nutrients[df_nutrients.pnns_groups_2 == 'Fats'].copy()

# Calculer l'énergie sans la contribution des glucides (~ énergie provenant des lipides)
df_fats[
    'energy_fats_100g'] = df_fats.energy_100g - slope_carbohydrates * df_fats.carbohydrates_100g

# Calculer la régression linéaire pour les produits gras uniquement
# entre l'énergie (sans la contribution des glucides) et les lipides.
slope_fats, intercept_fats, r_value, p_value, std_err = linregress(
    x=df_fats['fat_100g'], y=df_fats['energy_fats_100g'])

print("Équation de régression linéaire [kJ]    : Énergie = {} * Lipides + {}".
      format(slope_fats, intercept_fats))
print("Équation de régression linéaire [kcal]  : Énergie = {} * Lipides + {}".
      format(slope_fats * kj_to_kcal, intercept_fats * kj_to_kcal))
print("Énergie des Lipides [kJ]   apportent : {}".format(slope_fats))
print("Énergie des Lipides [kcal] apportent : {}".format(slope_fats *
                                                         kj_to_kcal))
```

1 gramme de graisses fournit environ 9 kilocalories d'énergie.

Nous pouvons maintenant déduire des légumineuses l'apport calorique des protéines

```python
# Nuage de points de l'énergie en fonction de chaque macronutriment pour les légumineuses uniquement.
pairplot_for_pnns_group(
    df_nutrients, 'Legumes', x_vars, y_vars,
    'Énergie par rapport aux macronutriments (légumineuses uniquement)')
```

Intéressant, nous pouvons voir différents clusters. Cela signifie que notre groupe PNNS de Légumineuses est trop large et pourrait être divisé en plusieurs sous-groupes plus spécifiques.

Nous pouvons deviner une corrélation linéaire pour les trois macronutriments.

```python
# Calculez la régression linéaire pour les fruits uniquement entre l'énergie et les glucides.
df_legumes = df_nutrients[df_nutrients.pnns_groups_2 == 'Legumes'].copy()

# Calculez l'énergie sans la contribution des glucides.
df_legumes['energy_proteins_100g'] = df_legumes.energy_100g \
    - slope_carbohydrates * df_legumes.carbohydrates_100g \
    - slope_fats * df_legumes.fat_100g
```

Avant de calculer une régression linéaire, nous devons vérifier s'il existe réellement une corrélation linéaire entre les protéines et l'énergie que nous venons de calculer

```python
# Tracer un nuage de points de l'énergie des protéines par rapport aux protéines.
sns.lmplot(data=df_legumes,
           x='proteins_100g',
           y='energy_proteins_100g',
           fit_reg=False,
           scatter_kws={
               "marker": "D",
               "s": 20
           },
           size=8)
plt.xlim(0, 100)
plt.xlabel('Poids des protéines (gr)')
plt.ylabel('Énergie provenant des protéines (calculée)')
plt.title(
    "Énergie calculée des protéines par rapport au poids des protéines (Légumineuses uniquement)"
)
plt.show()
```

```python
# Calculer la régression linéaire pour les légumineuses uniquement entre
# l'énergie (sans les apports en glucides et lipides) et les protéines
slope_proteins, intercept_proteins, r_value, p_value, std_err = linregress(
    x=df_legumes['proteins_100g'], y=df_legumes['energy_proteins_100g'])

print(
    "Équation de régression linéaire [kJ]    : Énergie = {} * Protéines + {}".
    format(slope_proteins, intercept_proteins))
print(
    "Équation de régression linéaire [kcal]  : Énergie = {} * Protéines + {}".
    format(slope_proteins * kj_to_kcal, intercept_proteins * kj_to_kcal))
print("Énergie des Protéines [kJ]   apportent : {}".format(slope_proteins))
print("Énergie des Protéines [kcal] apportent : {}".format(slope_proteins *
                                                           kj_to_kcal))
```

1 gramme de protéines fournit environ 5 kilocalories d'énergie.

Nous avons utilisé une méthode empirique pour découvrir le rapport calorique fourni par les glucides, les lipides et les protéines. Il est tentant de dire que la formule permettant de calculer l'énergie provenant du poids des trois macronutriments est la suivante :

    Énergie [kJ] = 16.898 * Glucides [g] + 18.179 * Graisses [g] + 20.210 * Protéines [g]

ou

    Énergie [kcal] = 4.038 * Glucides [g] + 8.659 * Graisses [g] + 4.830 * Protéines [g].

Et pour s'en souvenir plus facilement :

    Énergie [kcal] = 9 * Graisses [g] + 4 * Glucides [g] + 5 * Protéines [g] (9/4/5).
    
Note : d'après cette source https://www.cerin.org/glossaire/calorie/ le véritable ratio serait (9/4/4).

Nous pouvons évaluer la qualité de notre "modèle" très basique et empirique en calculant l'énergie prédite par le modèle pour chaque produit et en comparant l'écart type par rapport à l'énergie connue. Nous exclurons les groupes d'aliments utilisés pour construire notre "modèle".

```python
# Conserver uniquement les aliments non utilisés pendant le processus de modelage
df_model_test = df_nutrients[~df_nutrients.pnns_groups_2.
                             isin(['Fruits', 'Fats', 'Legumes'])].copy()

# Calculer l'énergie attendue
df_model_test['expected_energy_100g'] = slope_carbohydrates * df_model_test.carbohydrates_100g + \
    slope_fats * df_model_test.fat_100g + \
    slope_proteins * df_model_test.proteins_100g

# Calculer la valeur absolue de l'écart en pourcentage
df_model_test['energy_deviation'] = np.abs(
    df_model_test.expected_energy_100g -
    df_model_test.energy_100g) / df_model_test.energy_100g * 100
```

```python
# Obtenir des statistiques de l'écart type
df_model_test.energy_deviation.describe()
```

Avec une moyenne à environ 15% et une médiane à environ 2%, c'est correct pour un "modèle" empirique.
